### 93. Restore IP Addresses

Given a string s containing only digits, return all possible valid IP addresses that can be obtained from s. You can return them in any order.

A valid IP address consists of exactly four integers, each integer is between 0 and 255, separated by single dots and cannot have leading zeros. For example, "0.1.2.201" and "192.168.1.1" are valid IP addresses and "0.011.255.245", "192.168.1.312" and "192.168@1.1" are invalid IP addresses. 

#### Example 1:

- Input: s = "25525511135"
- Output: ["255.255.11.135","255.255.111.35"]

#### Example 2:

- Input: s = "0000"
- Output: ["0.0.0.0"]

#### Example 3:

- Input: s = "1111"
- Output: ["1.1.1.1"]

#### Example 4:

- Input: s = "010010"
- Output: ["0.10.0.10","0.100.1.0"]

#### Example 5:

- Input: s = "101023"
- Output: ["1.0.10.23","1.0.102.3","10.1.0.23","10.10.2.3","101.0.2.3"]
 
#### Constraints:

- 0 <= s.length <= 3000
- s consists of digits only.


```python
class Solution:
    def restoreIpAddresses(self, s: str) -> List[str]:
        res = []
        # Here, the list with length of 4 is initialized, and the corresponding 4 fragments are stored respectively
        sub_res = [0] * 4

        def dfs(seg_id, seg_first):
            # Let's look at the four clips
            # It is possible that characters are fully used and not fully used
            if seg_id == 4:
                # Use complete to add to the list and return
                if seg_first == len(s):
                    res.append('.'.join(str(seg)for seg in sub_res))
                # If it is not fully used, it will be returned directly
                return
            
            # If four fragments are not found, continue searching
            # Note that the characters are used up when four fragments are not found
            if seg_first == len(s):
                return

            # There cannot be a leading 0. If there is a 0, the current fragment can only be a single 0
            if s[seg_first] == "0":
                sub_res[seg_id] = 0
                dfs(seg_id+1, seg_first+1)

            addr = 0
            # Each clip selection
            for seg_last in range(seg_first, len(s)):
                # Here we use integer to express, and then convert to string type to avoid too frequent slicing
                addr = addr * 10 + (ord(s[seg_last]) - ord('0'))

                # Greater than 0 but less than or equal to 255
                if 0 < addr <= 255:
                    sub_res[seg_id] = addr
                    dfs(seg_id+1, seg_last+1)
                # Other situations jump out of the loop directly
                else:
                    break
            
        dfs(0, 0)
        return res  
```
